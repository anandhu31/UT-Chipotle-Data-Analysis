# Visualizing Chipotle Orders of UT students

This page will explain the steps taken to obtain, visualize, and compare UT students orders at Chipotle. The following questions are the attributes for this data: White Rice or Brown Rice, Black Beans or Pinto Beans, Meat or Veggie, Guac or No Guac. 

The data was obtained by asking people walking on Guadalupe Street 4 questions. This data was logged and calculated on Google Sheets. 

#####  Libraries used for charting the data

```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

### Charting for preference of Ingredients

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(10,4))  # 1 row, 2 columns
fig.subplots_adjust(wspace=0.5)

#rices
sizes = [48,52]
labels = 'Brown rice','White rice'

ax1.pie(sizes, labels = labels, autopct = '%.1f%%')
ax1.set_title('Brown rice vs. White rice')
ax1.axis('equal')

#beans
sizes = [32,68]
labels = 'Pinto Beans','Black Beans'

ax2.pie(sizes, labels = labels, autopct = '%.1f%%')
ax2.set_title('Pinto vs. Black beans')
ax2.axis('equal')

#meat
sizes = [48,52]
labels = 'Veggie','Meat'

ax3.pie(sizes, labels = labels, autopct = '%.1f%%')
ax3.set_title('Veggies vs. Meat')
ax3.axis('equal')

#guac
sizes = [40,60]
labels = 'No Guac','Guac'

ax4.pie(sizes, labels = labels, autopct = '%.1f%%')
ax4.set_title('No Guac vs. Guac')
ax4.axis('equal')
```

The Pie charts above show results from DAY 1 of the survey. DAY 1 contains the first 50 participants of the survery. Results show that the most combination of a Chipotle order is White rice, Black beans, Meat, and No Guac. The results for DAY 2 (next 50 participants) will be shown below.

```python
fig, (ax1,ax2,ax3,ax4) = plt.subplots(1,4,figsize=(10,4))  # 1 row, 2 columns
fig.subplots_adjust(wspace=0.5)

#rices
sizes = [42,58]
labels = 'Brown rice','White rice'

ax1.pie(sizes, labels = labels, autopct = '%.1f%%')
ax1.set_title('Brown rice vs. White rice')
ax1.axis('equal')

#beans
sizes = [38,62]
labels = 'Pinto Beans','Black Beans'

ax2.pie(sizes, labels = labels, autopct = '%.1f%%')
ax2.set_title('Pinto vs. Black beans')
ax2.axis('equal')

#meat
sizes = [48,52]
labels = 'Veggie','Meat'

ax3.pie(sizes, labels = labels, autopct = '%.1f%%')
ax3.set_title('Veggies vs. Meat')
ax3.axis('equal')

#guac
sizes = [40,60]
labels = 'No Guac','Guac'

ax4.pie(sizes, labels = labels, autopct = '%.1f%%')
ax4.set_title('No Guac vs. Guac')
ax4.axis('equal')
```

The results from DAY 1 and DAY 2 are very similar as the most popular order remains the same. 

## Conclusion

Therefore, the most popular order for people walking on Gaudalupe Street is White rice, Black beans, Meat, and No Guac
